# Question 2: which city has the maximum number of sales ?


```python
#Lets add a column of a city 
#Think like this{Lets create a packet which is called x
#which is called x which is carried by a bike and here the bike is the function that we create
#the function[read the bike] carries the packet and takes to another place
#In the other place we do an operation called splitting the string
#The string was carried inside a variable called x 
def get_the_city(address):
    return address.split(',')[1]


def get_state(address):
    return address.split(',')[2].split(' ')[1]



all_data["City"]=all_data["Purchase Address"].apply(lambda x:get_the_city(x)+'('+get_state(x)+')')
all_data.head()
result=all_data.groupby("City")["Sales"].sum()
result


```

```python
import matplotlib.ticker as ticker



cities=all_data["City"].unique()
array1=np.array(result)
plt.bar(cities,array1)
plt.xticks(cities,rotation='vertical',size=8)
plt.ylabel('Sales')
plt.xlabel('Cities')
plt.title("Sales along with cities")
```

# What products are more often sold to a particular id?

# We need to do what?
  *To keep Only these order id's that ordered more than once
  *Removing rest of the rows,So definately we need to create another data frame for this
  

```python



all_data.head()
data_frame=all_data[all_data["Order ID"].duplicated(keep=False)]  
#The .duplicated(keep=False) method checks for duplicate values in the "Order ID" column.
#If an "Order ID" appears more than once, it marks all occurrences as True.
#If an "Order ID" appears only once, it is marked as False.
data_frame.head(20)


```

# Now we need to groupby the data

```python


data_frame["Grouped"]=data_frame.groupby("Order ID")['Product'].transform(lambda x:','.join(x))    

#We cant use .sum() in this case because in this case we need to sum up some strings 
#To sum up strings there are no functions 
#So we need to create a custom function which will to join these strings 
#The function will be called join 
data_frame.head(5)
#Lets make a duplicate of Order id and Grouped Data to another dataframe 
data_frame=data_frame[["Order ID","Grouped"]].drop_duplicates()

data_frame.head()
```

```python
data_frame.count()

#We need to count the same product series purshed by different user ids 
from itertools import combinations 
from collections import Counter
count=Counter()



count=Counter()


for i in data_frame["Grouped"]:
    row_list=i.split(',')
    count.update(Counter(combinations(row_list,2)))  #Which two items sold the most
print(count)
count.most_common(10)

```
